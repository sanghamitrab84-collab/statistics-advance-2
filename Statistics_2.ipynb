
Question 1: What is hypothesis testing in statistics?
ANSWER:Hypothesis testing is a statistical method used to make decisions or inferences about a population parameter based on sample data. It provides a systematic procedure to test whether a claim (hypothesis) about a population is likely to be true or false.
Key Concepts:

Hypothesis:

Null Hypothesis (H₀): Assumes there is no effect or no difference. It is the default assumption.

Alternative Hypothesis (H₁ or Ha): Represents what the researcher wants to prove; it suggests that there is an effect or a difference.

Types of Errors:

Type I Error (α): Rejecting H₀ when it is actually true (false positive).

Type II Error (β): Failing to reject H₀ when it is actually false (false negative).

Level of Significance (α):
The probability of making a Type I error, often set at 0.05 (5%).

Test Statistic:
A numerical value (e.g., z, t, χ², F) calculated from sample data to compare against a critical value.

Decision Rule:

If the test statistic falls in the critical region, reject H₀.

Otherwise, fail to reject H₀.

p-value Approach:

If p-value ≤ α, reject H₀.

If p-value > α, fail to reject H₀.

Steps in Hypothesis Testing:

State the null (H₀) and alternative (H₁) hypotheses.

Select the level of significance (α).

Choose the appropriate statistical test.

Calculate the test statistic using sample data.

Determine the critical value or p-value.

Make a decision: reject or fail to reject H₀.

Draw a conclusion in context of the problem.

Example:
Suppose a company claims that the average lifetime of its batteries is 100 hours.

H₀: μ = 100 (no difference)

H₁: μ ≠ 100 (difference exists)
By collecting a sample and performing a t-test, the company can decide whether to accept or reject its claim with a certain level of confidence.

Conclusion:
Hypothesis testing is a fundamental tool in inferential statistics. It helps researchers and decision-makers draw reliable conclusions about populations from limited sample data while controlling for uncertainty and error.

Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?
ANSWER:1. Definition of Null Hypothesis (H₀):
The null hypothesis is a formal statement in hypothesis testing that assumes no effect, no difference, or no relationship exists in the population. It serves as the default or starting assumption that researchers try to test against.

Example: The average height of students is 165 cm (H₀: μ = 165).

2. Definition of Alternative Hypothesis (H₁ or Ha):
The alternative hypothesis is a statement that contradicts the null hypothesis. It suggests that there is an effect, a difference, or a relationship in the population. This is usually what the researcher aims to prove.

Example: The average height of students is not 165 cm (H₁: μ ≠ 165).

3. Relationship Between H₀ and H₁:

They are mutually exclusive: if one is true, the other must be false.

They are exhaustive: together they cover all possible outcomes.

4. Example in Context:
A drug manufacturer claims that a new medicine has no effect on blood pressure.

H₀: The drug has no effect (mean change = 0).

H₁: The drug has an effect (mean change ≠ 0).
By collecting sample data and performing a statistical test, we can decide whether to reject H₀ in favor of H₁.

Conclusion:
The null hypothesis (H₀) represents the default assumption of no difference, while the alternative hypothesis (H₁) represents the researcher’s claim of a difference or effect. The process of hypothesis testing revolves around determining whether there is enough evidence to reject H₀ in favor of H₁.

Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
ANSWER:1. Definition of Significance Level (α):
The significance level (α) is the probability of rejecting the null hypothesis (H₀) when it is actually true. In other words, it represents the risk of committing a Type I error. It is chosen by the researcher before conducting the test.

Common choices are α = 0.05 (5%), α = 0.01 (1%), or α = 0.10 (10%).

2. Role in Hypothesis Testing:

Sets the Threshold for Decision-Making:

The significance level defines the boundary between the acceptance and rejection regions for H₀.

Determines the Critical Value(s):

Based on α and the type of test (z, t, χ², F), we find the critical region in the probability distribution.

Example: For α = 0.05 in a two-tailed z-test, the critical values are ±1.96.

Controls Type I Error:

If α = 0.05, it means there is a 5% risk of rejecting H₀ when it is true.

Smaller α reduces this risk but makes it harder to reject H₀.

Used in p-value Comparison:

If p-value ≤ α, reject H₀.

If p-value > α, fail to reject H₀.

3. Example:
Suppose a company claims that the mean life of its bulbs is 1000 hours.

H₀: μ = 1000

H₁: μ ≠ 1000

α = 0.05

After testing, if the calculated p-value = 0.03, since 0.03 < 0.05, we reject H₀ and conclude the mean life is significantly different from 1000 hours.

4. Key Points About Significance Level:

It is pre-determined before the test.

Lower α (e.g., 0.01) = more strict test (reduces false positives).

Higher α (e.g., 0.10) = more lenient test (increases chance of rejecting H₀).

Choice depends on context:

Medical research → α = 0.01 (minimize risk of false claims).

Social science → α = 0.05 or 0.10 (more tolerance).

5. Conclusion:
The significance level (α) is a vital component of hypothesis testing. It defines the risk of error a researcher is willing to take, sets the rejection criteria, and plays a central role in interpreting results through critical values or p-values.

Question 4: What are Type I and Type II errors? Give examples of each.
ANSWER:1. Definition of Errors in Hypothesis Testing:
When making decisions in hypothesis testing, two types of errors can occur due to sampling variation. These are Type I error and Type II error.

2. Type I Error (α error):

Definition: Rejecting the null hypothesis (H₀) when it is actually true.

Also called: False Positive.

Probability of occurrence: Equal to the significance level (α).

Example:

A medical test concludes a patient has a disease when in fact the patient is healthy.

In court: Convicting an innocent person.

3. Type II Error (β error):

Definition: Failing to reject the null hypothesis (H₀) when it is actually false.

Also called: False Negative.

Probability of occurrence: Denoted by β. The power of a test is (1 – β).

Example:

A medical test concludes a patient is healthy when in fact the patient has the disease.

In court: Letting a guilty person go free.
4. Diagram (for quick sketch in exams):

Draw two overlapping curves (H₀ true vs H₁ true).

Shade α region (Type I error) in the tail(s).

Shade β region (Type II error) under H₁ curve.

5. Conclusion:
Both Type I and Type II errors represent incorrect decisions in hypothesis testing.

Type I error (α) = false alarm.

Type II error (β) = missed detection.
A good statistical test aims to minimize both, though reducing one often increases the other.


Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.
ANSWER:1. Introduction:
Both the Z-test and the T-test are statistical methods used to test hypotheses about population parameters based on sample data. They are similar in purpose but differ in the assumptions and conditions under which they are applied.

2. Z-Test:

Definition: A statistical test based on the standard normal distribution (Z-distribution).

Conditions for Use:

Sample size is large (n ≥ 30).

Population standard deviation (σ) is known.

Data follows a normal distribution or approximately normal (by Central Limit Theorem for large n).

Applications:

Testing population mean (one-sample, two-sample).

Testing proportions.

Example: Testing if the mean height of 100 students differs from 165 cm, when σ is known.

3. T-Test:

Definition: A statistical test based on the Student’s t-distribution.

Conditions for Use:

Sample size is small (n < 30).

Population standard deviation (σ) is unknown (sample standard deviation s is used).

Data should be approximately normally distributed.

Types of T-tests:

One-sample t-test.

Independent two-sample t-test.

Paired t-test.

Example: Testing if the average marks of 15 students differ from 50 when σ is unknown.
Conclusion:

Use a Z-test when the sample size is large and σ is known.

Use a T-test when the sample size is small and σ is unknown.
Both tests help in making decisions about population means and are fundamental tools in inferential statistics.



Question 6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.
ANSWER:Concept:

A binomial distribution gives the probability of getting a fixed number of successes in a given number of independent Bernoulli trials.

Parameters:

n = 10 (number of trials)

p = 0.5 (probability of success)

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10       # number of trials
p = 0.5      # probability of success
size = 1000  # number of samples to generate

# Generate binomial distribution
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=np.arange(n+2)-0.5, edgecolor='black', density=True)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

ANSWER:1. Problem Setup (Hypotheses):
We want to test whether the sample mean differs from a hypothesized population mean (μ₀ = 50).

Null Hypothesis (H₀): μ = 50 (no difference)

Alternative Hypothesis (H₁): μ ≠ 50 (difference exists)

Significance Level (α): 0.05 (5%)

```python
import numpy as np
from scipy import stats

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Hypothesized population mean
mu_0 = 50

# Sample statistics
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)   # sample standard deviation
n = len(sample_data)

# Z-statistic calculation
z_stat = (sample_mean - mu_0) / (sample_std / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

# Output
print(f"Sample Mean = {sample_mean:.3f}")
print(f"Sample Std Dev = {sample_std:.3f}")
print(f"Z-statistic = {z_stat:.3f}")
print(f"P-value = {p_value:.4f}")
```

Question 8: Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
(Include your Python code and output in the code box below.)
ANSWER:1. Concept:

A confidence interval (CI) gives a range of values within which the true population mean is likely to lie with a given level of confidence (95% here).

Formula for CI:
x±Zα/2x s/(⎷n)





where

ˉ
x = sample mean,

s = sample standard deviation,

n = sample size,

0.025=1.96
Z0.025=1.96 for 95% CI.

```python

import matplotlib.pyplot as plt
from scipy import stats

# Simulate normal distribution data
np.random.seed(42)  # for reproducibility
data = np.random.normal(loc=50, scale=5, size=200)  # mean=50, std=5, n=200

# Sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)

# 95% Confidence Interval
alpha = 0.05
z_critical = stats.norm.ppf(1 - alpha/2)  # 1.96
margin_of_error = z_critical * (sample_std / np.sqrt(n))
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"Sample Mean = {sample_mean:.3f}")
print(f"95% Confidence Interval = ({ci_lower:.3f}, {ci_upper:.3f})")

# Plot histogram of data
plt.hist(data, bins=20, edgecolor='black', alpha=0.7, density=True)
plt.axvline(sample_mean, color='red', linestyle='--', label=f"Mean = {sample_mean:.2f}")
plt.axvline(ci_lower, color='green', linestyle='--', label=f"95% CI Lower = {ci_lower:.2f}")
plt.axvline(ci_upper, color='blue', linestyle='--', label=f"95% CI Upper = {ci_upper:.2f}")
plt.title("Normal Distribution with 95% Confidence Interval")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean,
ANSWER:1. Concept of Z-scores:

A Z-score indicates how many standard deviations a data point is away from the mean.

Formula:Z=(X-X')/s
where

X = data value,
ˉ
X= sample mean,
s = sample standard deviation.

Interpretation:

Z = 0 → data point equals the mean.

Z = +1 → one standard deviation above mean.

Z = –1 → one standard deviation below mean.

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate Z-scores
def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # sample standard deviation
    z_scores = [(x - mean) / std for x in data]
    return np.array(z_scores)

# Example dataset
data = [45, 48, 50, 52, 53, 47, 49, 51, 46, 54]

# Calculate Z-scores
z_scores = calculate_z_scores(data)

# Print results
print("Original Data:", data)
print("Z-scores:", np.round(z_scores, 2))

# Plot histogram of standardized data
plt.hist(z_scores, bins=7, edgecolor='black', alpha=0.7, density=True)
plt.title("Histogram of Z-scores (Standardized Data)")
plt.xlabel("Z-score")
plt.ylabel("Density")
plt.axvline(0, color='red', linestyle='--', label="Mean (Z=0)")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```
